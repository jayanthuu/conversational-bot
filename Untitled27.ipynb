```python
# Cell 3: Load BLIP model for image captioning
from transformers import BlipProcessor, BlipForConditionalGeneration
import warnings
warnings.filterwarnings('ignore')

print("⏳ Loading BLIP model (this may take 1-2 minutes)...")

# Load processor and model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Move model to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)

print(f"✅ BLIP model loaded successfully on {device}!")
print("Model is ready for image captioning.")
```

```python
# Cell 4: Function to generate caption from image
def generate_caption(image_path):
    """
    Generate a descriptive caption for an image using BLIP model.
    """
    try:
        # Open and convert image
        raw_image = Image.open(image_path).convert('RGB')

        # Process image for BLIP
        inputs = processor(raw_image, return_tensors="pt").to(device)

        # Generate caption
        with torch.no_grad():
            out = model.generate(**inputs, max_length=50, num_beams=5)

        caption = processor.decode(out[0], skip_special_tokens=True)
        return caption

    except Exception as e:
        return f"Error processing image: {str(e)}"

# Test with a sample image from internet (or upload your own)
print("📷 Testing image captioning with a sample image...")

# Download a sample image (a cat image from Wikimedia)
!wget -q -O test_image.jpg "https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Cat03.jpg/640px-Cat03.jpg"

# Display the test image
from IPython.display import Image as IPImage
display(IPImage(filename='test_image.jpg', width=300))

# Generate caption
caption = generate_caption("test_image.jpg")
print(f"\n📝 Generated Caption: {caption}")

print("\n✅ Image captioning test complete!")
```

```python
# Cell 5: Function to extract frames from video every 5 seconds
import cv2
import os
from google.colab import files

def extract_frames_from_video(video_path, interval_seconds=5):
    """
    Extract frames from video at specified interval.
    Returns list of frame images and their timestamps.
    """
    frames = []
    timestamps = []

    # Open video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"❌ Error: Could not open video {video_path}")
        return [], []

    # Get video properties
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / fps if fps > 0 else 0

    print(f"📹 Video Info: {fps:.1f} FPS, {total_frames} frames, {duration:.1f} seconds")

    # Calculate frame interval (every 5 seconds)
    frame_interval = int(fps * interval_seconds)

    frame_count = 0
    extracted_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Extract frame at interval
        if frame_count % frame_interval == 0:
            # Convert BGR (OpenCV) to RGB (PIL)
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frame_pil = Image.fromarray(frame_rgb)

            frames.append(frame_pil)
            timestamps.append(frame_count / fps)
            extracted_count += 1

        frame_count += 1

    cap.release()

    print(f"✅ Extracted {extracted_count} frames (every {interval_seconds} seconds)")
    return frames, timestamps

def upload_video():
    """Upload a video file in Colab"""
    print("📤 Please upload a video file (mp4, avi, mov)...")
    uploaded = files.upload()

    if not uploaded:
        print("❌ No file uploaded.")
        return None

    # Get the uploaded filename
    video_filename = list(uploaded.keys())[0]
    print(f"✅ Uploaded: {video_filename} ({uploaded[video_filename] / 1024:.1f} KB)")
    return video_filename

# Test with a sample video (optional - you can upload your own later)
print("Video processing functions defined successfully!")
print("\nWhen you're ready to test with your own video:")
print("1. Call upload_video() to upload")
print("2. Call extract_frames_from_video('your_video.mp4')")
```

```python
# Cell 7: Setup chatbot interface and conversation functions
import json
from datetime import datetime

class ImageChatbot:
    def __init__(self):
        self.conversation_history = []
        self.current_frames = []
        self.frame_timestamps = []
        self.frame_captions = []

    def set_video_frames(self, frames, timestamps):
        """Set frames extracted from video"""
        self.current_frames = frames
        self.frame_timestamps = timestamps
        self.frame_captions = []

        # Generate captions for all frames
        print(f"🔄 Generating captions for {len(frames)} frames...")
        for i, frame in enumerate(frames):
            # Save frame temporarily
            frame.save(f"frame_{i}.jpg")
            caption = generate_caption(f"frame_{i}.jpg")
            self.frame_captions.append(caption)
            # Clean up
            if os.path.exists(f"frame_{i}.jpg"):
                os.remove(f"frame_{i}.jpg")

        print("✅ All frame captions generated!")
        return self.frame_captions

    def get_context_summary(self):
        """Create a summary of video context from all frames"""
        if not self.frame_captions:
            return "No video context available."

        context = "Video Context (frames every 5 seconds):\n"
        for i, (caption, timestamp) in enumerate(zip(self.frame_captions, self.frame_timestamps)):
            context += f"  • At {timestamp:.1f}s: {caption}\n"
        return context

    def answer_question(self, user_question):
        """Answer user question based on video context"""
        if not self.frame_captions:
            return "Please process a video first to get context."

        # Get context summary
        context = self.get_context_summary()

        # Simple rule-based responses (we'll enhance this with LLM later)
        question_lower = user_question.lower()

        if "what" in question_lower and "see" in question_lower:
            return f"Based on the video analysis:\n{context}"

        elif "how many" in question_lower:
            return f"I analyzed {len(self.frame_captions)} frames from the video."

        elif "first" in question_lower:
            if self.frame_captions:
                return f"The first frame shows: {self.frame_captions[0]}"

        elif "last" in question_lower:
            if self.frame_captions:
                return f"The last frame shows: {self.frame_captions[-1]}"

        elif "describe" in question_lower:
            return f"Video Description:\n{context}"

        # Default response
        return f"I've analyzed the video. {context}\nYou asked: '{user_question}'"

# Initialize chatbot
chatbot = ImageChatbot()

print("✅ Chatbot class initialized successfully!")
print("\nReady to use:")
print("1. Upload video with upload_video()")
print("2. Extract frames with extract_frames_from_video()")
print("3. Set frames in chatbot: chatbot.set_video_frames(frames, timestamps)")
print("4. Ask questions: chatbot.answer_question('your question')")
```

```python
# Cell 10: REAL-TIME Video Analysis Interface - FINAL CORRECTED VERSION
import os
import time
import json
from IPython.display import display, HTML
from google.colab import files, output
import base64
import cv2
from PIL import Image
import numpy as np
import torch
from transformers import BlipProcessor, BlipForConditionalGeneration, pipeline
import tempfile
import io

print("🚀 Building REAL-TIME Video Analysis Interface...")

# ========== LOAD AI MODELS ==========
print("🔧 Loading AI models for real-time analysis...")

processor_global = None
model_global = None
dialogpt_model_global = None
device_global = "cuda" if torch.cuda.is_available() else "cpu"

# Load BLIP model
print("⏳ Ensuring BLIP models are loaded...")
try:
    processor_global = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
    model_global = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
    model_global.to(device_global)
    print(f"✅ BLIP model ready on {device_global}!")
except Exception as e:
    print(f"❌ Failed to load BLIP model: {e}")

# Load DialoGPT model
print("⏳ Ensuring DialoGPT model is loaded...")
try:
    dialogpt_model_global = pipeline(
        "text-generation",
        model="microsoft/DialoGPT-small",
        torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
        device=0 if torch.cuda.is_available() else -1
    )
    print("✅ DialoGPT model ready!")
except Exception as e:
    print(f"❌ Failed to load DialoGPT: {e}")

# Helper functions for JavaScript communication
def send_progress(percent, text):
    payload = {'name': 'update_progress', 'payload': {'percent': percent, 'text': text}}
    output.eval_js(f"google.colab.kernel.dispatch_frontend({{type: 'custom', data: {json.dumps(payload)}}})")

def send_analysis_result(status, data=None, error=None):
    payload = {'name': 'analysis_results', 'payload': {'status': status, 'data': data, 'error': error}}
    output.eval_js(f"google.colab.kernel.dispatch_frontend({{type: 'custom', data: {json.dumps(payload)}}})")

def send_llm_response(answer):
    payload = {'name': 'llm_response', 'payload': {'answer': answer}}
    output.eval_js(f"google.colab.kernel.dispatch_frontend({{type: 'custom', data: {json.dumps(payload)}}})")

# ========== REAL-TIME ANALYSIS FUNCTIONS ==========
def analyze_video_realtime(video_path):
    """Analyze video frame-by-frame and generate captions"""
    print(f"🔍 Starting REAL analysis of: {video_path}")

    frames_pil = []
    frames_base64 = []
    timestamps = []
    captions = []
    video_duration = 0.0

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return frames_base64, timestamps, captions, "Error: Could not open video", 0.0

    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    video_duration = total_frames / fps if fps > 0 else 0.0

    if fps <= 0:
        fps = 30

    frame_interval = int(fps * 5)  # Every 5 seconds
    if frame_interval == 0:
        frame_interval = 1

    frame_count = 0
    extracted_count = 0

    send_progress(10, "Extracting frames...")
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if frame_count % frame_interval == 0:
            try:
                frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                pil_image = Image.fromarray(frame_rgb)
                frames_pil.append(pil_image)
                timestamps.append(frame_count / fps)

                # Convert PIL image to base64 for display in JS
                buffered = io.BytesIO()
                pil_image.save(buffered, format="JPEG", quality=85)
                frames_base64.append(base64.b64encode(buffered.getvalue()).decode('utf-8'))

                extracted_count += 1
            except Exception as e:
                print(f"Warning: Could not process frame {frame_count}: {e}")

        frame_count += 1
        if frame_count > 1000 * fps:
            print(f"⚠️ Limiting frame processing to first {int(1000 * fps)} frames.")
            break

    cap.release()

    # Generate captions with BLIP
    if frames_pil and processor_global and model_global:
        send_progress(40, f"Analyzing {len(frames_pil)} frames with BLIP AI...")
        for i, frame in enumerate(frames_pil):
            try:
                inputs = processor_global(frame, return_tensors="pt").to(device_global)
                with torch.no_grad():
                    out = model_global.generate(**inputs, max_length=50, num_beams=5)
                caption = processor_global.decode(out[0], skip_special_tokens=True)
                captions.append(caption)
                progress = 40 + int((i+1)/len(frames_pil) * 50)
                send_progress(progress, f"Generating caption for frame {i+1}...")
            except Exception as e:
                captions.append(f"Frame {i+1}: Analysis error")
                print(f"Error captioning frame {i+1}: {e}")
    elif not frames_pil:
        print("No frames extracted for captioning.")
    else:
        print("BLIP models not loaded, skipping caption generation.")

    send_progress(95, "Finalizing analysis...")
    return frames_base64, timestamps, captions, "Success", video_duration

def answer_question_realtime(question, captions, timestamps):
    """Answer questions based on video analysis"""
    if not captions:
        return "No video analyzed yet. Please upload a video first."

    # Create context from analysis
    context = "Video Analysis Results:\n"
    for i, (caption, ts) in enumerate(zip(captions, timestamps)):
        context += f"• {ts:.1f}s: {caption}\n"

    # Try to use DialoGPT
    if dialogpt_model_global:
        prompt = f"""Based on this video analysis, answer the question concisely.

ANALYSIS:
{context}

QUESTION: {question}

ANSWER:"""

        try:
            response = dialogpt_model_global(
                prompt,
                max_length=200,
                num_return_sequences=1,
                temperature=0.7,
                do_sample=True,
                pad_token_id=dialogpt_model_global.tokenizer.eos_token_id
            )
            answer = response[0]['generated_text']
            if "ANSWER:" in answer:
                answer = answer.split("ANSWER:")[-1].strip()
            elif "Answer:" in answer:
                answer = answer.split("Answer:", 1)[-1].strip()
            return answer
        except Exception as e:
            print(f"⚠️ LLM generation failed: {e}. Falling back to rule-based.")

    # Fallback to rule-based
    q_lower = question.lower()
    if "what" in q_lower and ("see" in q_lower or "video" in q_lower):
        return f"Here's what I found in the video:\n{context}"
    elif "how many" in q_lower:
        return f"I analyzed {len(captions)} frames from your video."
    elif "first" in q_lower:
        return f"The first frame shows: {captions[0] if captions else 'No frames'}"
    elif "last" in q_lower:
        return f"The last frame shows: {captions[-1] if captions else 'No frames'}"
    elif "describe" in q_lower:
        return f"Video Description:\n{context}"
    else:
        return f"Based on my analysis: {context}\nYou asked: '{question}'"

# ========== CREATE INTERACTIVE INTERFACE ==========
print("\n🎨 Creating REAL-TIME interactive interface...")

interface_html = '''
<div id="realtime-chatbot" style="width: 100%; background: white; border-radius: 15px; padding: 25px; box-shadow: 0 10px 30px rgba(0,0,0,0.1); font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;">

    <!-- Header -->
    <div style="text-align: center; padding: 25px; background: linear-gradient(135deg, #1E88E5 0%, #0D47A1 100%); border-radius: 12px; color: white; margin-bottom: 25px;">
        <h1 style="margin: 0; font-size: 32px; font-weight: 600;">🎥 REAL-TIME Video Understanding Chatbot</h1>
        <p style="margin: 10px 0 0 0; opacity: 0.9; font-size: 16px;">Upload ANY video → REAL AI analysis → Ask questions in real-time</p>
    </div>

    <!-- Main Content -->
    <div style="display: flex; gap: 25px;">
        <!-- Left Panel: Video Upload & Control -->
        <div style="flex: 1; background: #f8faff; padding: 25px; border-radius: 12px; border: 2px solid #e3f2fd;">
            <h3 style="color: #1E88E5; margin-top: 0; font-size: 22px;">📤 Upload & Analyze</h3>
            <div id="upload-area" style="text-align: center; padding: 30px; border: 3px dashed #90caf9; border-radius: 12px; background: white; margin-bottom: 20px;">
                <div style="font-size: 60px; color: #1E88E5; margin-bottom: 15px;">📹</div>
                <h4 style="color: #333; margin: 0 0 10px 0;">Drag & drop your video here</h4>
                <p style="color: #666; margin-bottom: 20px; font-size: 14px;">Or click below to browse files</p>
                <button id="upload-btn" style="background: linear-gradient(to right, #1E88E5, #1565C0); color: white; border: none; padding: 14px 28px; border-radius: 8px; cursor: pointer; font-size: 16px; font-weight: 600;">
                    📁 Choose Video File
                </button>
                <input type="file" id="video-input" accept="video/*" style="display: none;">
                <p style="font-size: 13px; color: #78909C; margin-top: 15px;">Supports: MP4, AVI, MOV, MKV (max 100MB)</p>
            </div>

            <!-- Analysis Controls -->
            <div id="control-panel" style="display: none;">
                <h4 style="color: #1E88E5; margin-top: 25px; border-bottom: 2px solid #e3f2fd; padding-bottom: 10px;">🎬 Analysis Controls</h4>
                <div style="background: white; padding: 20px; border-radius: 10px; margin-top: 15px;">
                    <div style="display: flex; justify-content: space-between; margin-bottom: 15px;">
                        <span style="color: #555; font-weight: 500;">Selected Video:</span>
                        <span id="selected-video" style="color: #1E88E5; font-weight: 600;"></span>
                    </div>
                    <button id="analyze-btn" style="width: 100%; background: linear-gradient(to right, #4CAF50, #2E7D32); color: white; border: none; padding: 16px; border-radius: 8px; cursor: pointer; font-size: 16px; font-weight: 600; margin-top: 10px;">
                        🤖 Start REAL AI Analysis
                    </button>
                    <div style="display: flex; gap: 10px; margin-top: 15px;">
                        <button id="cancel-btn" style="flex: 1; background: #f5f5f5; color: #666; border: 1px solid #ddd; padding: 12px; border-radius: 6px; cursor: pointer;">❌ Cancel</button>
                        <button id="new-video-btn" style="flex: 1; background: #f5f5f5; color: #666; border: 1px solid #ddd; padding: 12px; border-radius: 6px; cursor: pointer;">📹 New Video</button>
                    </div>
                </div>
            </div>

            <!-- Analysis Progress -->
            <div id="progress-section" style="display: none; margin-top: 25px;">
                <h4 style="color: #1E88E5; border-bottom: 2px solid #e3f2fd; padding-bottom: 10px;">🔄 Analysis Progress</h4>
                <div style="background: white; padding: 20px; border-radius: 10px; margin-top: 15px;">
                    <div id="progress-text" style="color: #555; margin-bottom: 15px; font-size: 15px;">Initializing AI models...</div>
                    <div style="width: 100%; height: 12px; background: #e0e0e0; border-radius: 6px; overflow: hidden; margin-bottom: 8px;">
                        <div id="progress-bar" style="width: 0%; height: 100%; background: linear-gradient(to right, #1E88E5, #0D47A1); transition: width 0.5s;"></div>
                    </div>
                    <div style="display: flex; justify-content: space-between;">
                        <span id="progress-percent" style="font-size: 14px; color: #1E88E5; font-weight: 600;">0%</span>
                        <span id="progress-status" style="font-size: 14px; color: #666;">Waiting</span>
                    </div>
                </div>
            </div>
        </div>

        <!-- Right Panel: Chat Interface -->
        <div style="flex: 2;">
            <div style="display: flex; justify-content: space-between; align-items: center; margin-bottom: 20px;">
                <h3 style="color: #1E88E5; margin: 0; font-size: 22px;">💬 Chat with AI About Your Video</h3>
                <div style="background: #e8f5e9; color: #2E7D32; padding: 8px 15px; border-radius: 20px; font-size: 14px; font-weight: 600;">🤖 REAL AI Mode</div>
            </div>
            <div id="chat-display" style="height: 450px; overflow-y: auto; background: #fafafa; padding: 20px; border-radius: 12px; margin-bottom: 20px; border: 1px solid #e0e0e0;">
                <div id="welcome-message" style="text-align: center; padding: 40px 20px; color: #666;">
                    <div style="font-size: 48px; margin-bottom: 20px;">👋</div>
                    <h4 style="color: #1E88E5; margin-bottom: 15px;">Welcome to REAL-TIME Video Analysis!</h4>
                    <p>Upload any video and I'll analyze it with AI in real-time.</p>
                    <p>Then ask me anything about what's in your video!</p>
                </div>
            </div>
            <div id="chat-input-section" style="display: none;">
                <div style="display: flex; gap: 12px; margin-bottom: 10px;">
                    <input type="text" id="chat-input" placeholder="Ask anything about your video..." style="flex: 1; padding: 16px; border: 2px solid #ddd; border-radius: 8px; font-size: 15px;">
                    <button id="send-btn" style="background: linear-gradient(to right, #1E88E5, #1565C0); color: white; border: none; padding: 0 30px; border-radius: 8px; cursor: pointer; font-size: 16px; font-weight: 600; min-width: 100px;">Send</button>
                </div>
                <div style="display: flex; justify-content: space-between;">
                    <p style="font-size: 13px; color: #78909C; margin: 5px 0 0 0;">💡 Try: "What do you see?", "Describe frame at 10 seconds"</p>
                    <button id="clear-chat-btn" style="background: none; border: none; color: #f44336; cursor: pointer; font-size: 13px; padding: 5px 10px;">🗑️ Clear Chat</button>
                </div>
            </div>
        </div>
    </div>
</div>

<script>
// Global state
let currentVideo = null;
let analysisData = {
    frames: [],
    timestamps: [],
    captions: [],
    videoName: '',
    videoDuration: 0
};
let isAnalyzing = false;

// DOM Elements
const uploadBtn = document.getElementById('upload-btn');
const videoInput = document.getElementById('video-input');
const controlPanel = document.getElementById('control-panel');
const progressSection = document.getElementById('progress-section');
const analyzeBtn = document.getElementById('analyze-btn');
const chatInputSection = document.getElementById('chat-input-section');
const chatDisplay = document.getElementById('chat-display');
const chatInput = document.getElementById('chat-input');
const sendBtn = document.getElementById('send-btn');
const welcomeMessage = document.getElementById('welcome-message');

// Event Listeners
uploadBtn.addEventListener('click', () => videoInput.click());

videoInput.addEventListener('change', function(e) {
    if (this.files.length > 0) {
        handleVideoUpload(this.files[0]);
    }
});

analyzeBtn.addEventListener('click', startRealTimeAnalysis);
sendBtn.addEventListener('click', sendMessage);
chatInput.addEventListener('keypress', function(e) {
    if (e.key === 'Enter') sendMessage();
});

document.getElementById('clear-chat-btn').addEventListener('click', clearChat);
document.getElementById('cancel-btn').addEventListener('click', cancelAnalysis);
document.getElementById('new-video-btn').addEventListener('click', resetToUpload);

// Handle video upload
function handleVideoUpload(file) {
    if (!file.type.includes('video/')) {
        alert('Please upload a video file (MP4, AVI, MOV, etc.)');
        return;
    }
    if (file.size > 100 * 1024 * 1024) {
        alert('Video file is too large. Please upload a video under 100MB.');
        return;
    }
    currentVideo = file;
    document.getElementById('selected-video').textContent = file.name;
    document.getElementById('control-panel').style.display = 'block';
    addMessage('bot', `📹 Video selected: <b>${file.name}</b>`);
    addMessage('bot', `📊 File size: ${(file.size / (1024*1024)).toFixed(2)} MB`);
    addMessage('bot', 'Click "Start REAL AI Analysis" to begin processing with AI models.');
}

// Start real-time analysis
function startRealTimeAnalysis() {
    if (!currentVideo || isAnalyzing) return;
    isAnalyzing = true;
    progressSection.style.display = 'block';
    analyzeBtn.disabled = true;
    analyzeBtn.innerHTML = '⏳ Processing...';
    analyzeBtn.style.background = 'linear-gradient(to right, #FF9800, #F57C00)';

    const reader = new FileReader();
    reader.onload = function(event) {
        const base64Data = event.target.result.split(',')[1];
        google.colab.kernel.invokeFunction('analyze_video_from_js', [base64Data, currentVideo.name, currentVideo.type], {});
    };
    reader.readAsDataURL(currentVideo);
}

// ========== CRITICAL FIX ==========
// Changed "goggle" to "google" - This was the main bug!
google.colab.kernel.onResults((results) => {
    if (results.type === 'custom' && results.data) {
        const data = results.data.payload;

        if (results.data.name === 'analysis_results') {
            if (data.status === 'Success') {
                analysisData.frames = data.data.frames;
                analysisData.timestamps = data.data.timestamps;
                analysisData.captions = data.data.captions;
                analysisData.videoName = currentVideo.name;
                analysisData.videoDuration = data.data.video_duration;
                completeAnalysis();
            } else {
                console.error('Analysis failed:', data.error);
                addMessage('bot', `❌ Video analysis failed: ${data.error}`);
                finishProcessing();
            }
        } else if (results.data.name === 'llm_response') {
            const thinkingEl = document.getElementById('thinking-indicator');
            if (thinkingEl) thinkingEl.remove();
            addMessage('bot', data.answer);
        } else if (results.data.name === 'update_progress') {
            updateProgress(data.percent, data.text);
        }
    }
});

// Update progress bar
function updateProgress(percent, text) {
    document.getElementById('progress-bar').style.width = percent + '%';
    document.getElementById('progress-percent').textContent = percent + '%';
    document.getElementById('progress-text').textContent = text;
    document.getElementById('progress-status').textContent = text;
}

// Complete analysis
function completeAnalysis() {
    isAnalyzing = false;
    addMessage('bot', '✅ <b>REAL AI Analysis Complete!</b>');
    addMessage('bot', `I analyzed your video (${analysisData.videoName}) using BLIP vision AI and prepared conversational AI.`);
    addMessage('bot', 'You can now ask me anything about your video!');

    chatInputSection.style.display = 'block';
    welcomeMessage.style.display = 'none';

    setTimeout(() => {
        analyzeBtn.disabled = false;
        analyzeBtn.innerHTML = '✅ Analysis Complete';
        analyzeBtn.style.background = 'linear-gradient(to right, #4CAF50, #2E7D32)';
    }, 1000);
}

function finishProcessing() {
    isAnalyzing = false;
    analyzeBtn.disabled = false;
    analyzeBtn.innerHTML = '🤖 Start REAL AI Analysis';
    analyzeBtn.style.background = 'linear-gradient(to right, #4CAF50, #2E7D32)';
}

// Add message to chat
function addMessage(sender, text) {
    const messageDiv = document.createElement('div');
    messageDiv.style.cssText = `
        background: ${sender === 'user' ? '#E3F2FD' : '#FFFFFF'};
        padding: 18px 20px;
        margin: 15px 0;
        border-radius: 12px;
        border-left: 6px solid ${sender === 'user' ? '#1E88E5' : '#4CAF50'};
        box-shadow: 0 2px 8px rgba(0,0,0,0.05);
        ${sender === 'user' ? 'margin-left: 40px;' : 'margin-right: 40px;'}
    `;

    messageDiv.innerHTML = `
        <div style="display: flex; align-items: center; margin-bottom: 8px;">
            <div style="font-weight: 600; color: ${sender === 'user' ? '#1565C0' : '#2E7D32'};">${sender === 'user' ? '👤 You' : '🤖 Video AI Assistant'}</div>
            ${sender === 'bot' ? '<div style="margin-left: 10px; font-size: 11px; background: #E8F5E9; color: #2E7D32; padding: 2px 8px; border-radius: 10px;">REAL AI</div>' : ''}
        </div>
        <div style="color: #333; line-height: 1.5;">${text}</div>
    `;

    chatDisplay.appendChild(messageDiv);
    chatDisplay.scrollTop = chatDisplay.scrollHeight;
}

// Send message
function sendMessage() {
    const question = chatInput.value.trim();
    if (!question) return;

    if (!analysisData.frames.length) {
        addMessage('bot', 'Please process a video first to get REAL analysis!');
        return;
    }

    addMessage('user', question);
    chatInput.value = '';

    // Show thinking indicator
    const thinkingDiv = document.createElement('div');
    thinkingDiv.id = 'thinking-indicator';
    thinkingDiv.style.cssText = `
        background: #FFF3E0;
        padding: 15px 20px;
        margin: 15px 0;
        border-radius: 12px;
        border-left: 6px solid #FF9800;
        margin-right: 40px;
    `;
    thinkingDiv.innerHTML = `
        <div style="display: flex; align-items: center; gap: 10px;">
            <div class="thinking-dots"></div>
            <div style="font-weight: 600; color: #E65100;">Video AI is analyzing your question...</div>
        </div>
    `;

    chatDisplay.appendChild(thinkingDiv);
    chatDisplay.scrollTop = chatDisplay.scrollHeight;

    // Send question to Python backend
    google.colab.kernel.invokeFunction('answer_question_from_js',
        [question, analysisData.captions, analysisData.timestamps], {});
}

// Clear chat
function clearChat() {
    if (confirm('Clear all chat history?')) {
        chatDisplay.innerHTML = '';
        if (analysisData.videoName) {
            welcomeMessage.style.display = 'block';
            welcomeMessage.innerHTML = `
                <div style="font-size: 48px; margin-bottom: 20px;">✅</div>
                <h4 style="color: #1E88E5; margin-bottom: 15px;">Video Analysis Ready!</h4>
                <p>Your video "<b>${analysisData.videoName}</b>" has been analyzed.</p>
                <p>You can ask me anything about it!</p>
            `;
        }
    }
}

function cancelAnalysis() {
    if (confirm('Cancel current analysis?')) {
        isAnalyzing = false;
        progressSection.style.display = 'none';
        analyzeBtn.disabled = false;
        analyzeBtn.innerHTML = '🤖 Start REAL AI Analysis';
        analyzeBtn.style.background = 'linear-gradient(to right, #4CAF50, #2E7D32)';
        addMessage('bot', 'Analysis cancelled. You can upload a new video or restart analysis.');
    }
}

function resetToUpload() {
    if (confirm('Start over with a new video?')) {
        currentVideo = null;
        analysisData = { frames: [], timestamps: [], captions: [], videoName: '', videoDuration: 0 };
        controlPanel.style.display = 'none';
        progressSection.style.display = 'none';
        chatInputSection.style.display = 'none';
        videoInput.value = '';
        chatDisplay.innerHTML = '';
        welcomeMessage.style.display = 'block';
        welcomeMessage.innerHTML = `
            <div style="font-size: 48px; margin-bottom: 20px;">👋</div>
            <h4 style="color: #1E88E5; margin-bottom: 15px;">Welcome to REAL-TIME Video Analysis!</h4>
            <p>Upload any video and I'll analyze it with AI in real-time.</p>
            <p>Then ask me anything about what's in your video!</p>
        `;
        addMessage('bot', 'Ready for a new video upload! Click "Choose Video File" to begin.');
    }
}

// Add CSS animations
const style = document.createElement('style');
style.textContent = `
    .thinking-dots::after {
        content: '...';
        animation: dots 1.5s steps(4, end) infinite;
    }
    @keyframes dots {
        0%, 20% { content: '.'; }
        40% { content: '..'; }
        60%, 100% { content: '...'; }
    }
`;
document.head.appendChild(style);
</script>
'''

display(HTML(interface_html))

# Expose Python functions to JavaScript
def analyze_video_from_js(base64_data, filename, file_type):
    send_progress(5, f"Receiving video '{filename}'...")
    try:
        # Save base64 data to a temporary video file
        with tempfile.NamedTemporaryFile(delete=False, suffix=os.path.splitext(filename)[1]) as temp_video_file:
            temp_video_file.write(base64.b64decode(base64_data))
            temp_video_path = temp_video_file.name
        print(f"Saved temporary video to: {temp_video_path}")

        # Perform analysis
        frames_base64, timestamps, captions, status_message, video_duration = analyze_video_realtime(temp_video_path)

        # Clean up temporary file
        os.remove(temp_video_path)

        if "Error" in status_message:
            send_analysis_result('error', error=status_message)
        else:
            results = {
                'frames': frames_base64,
                'timestamps': timestamps,
                'captions': captions,
                'video_duration': video_duration
            }
            send_analysis_result('Success', data=results)

    except Exception as e:
        print(f"Error during video analysis: {e}")
        send_analysis_result('error', error=str(e))

    send_progress(100, "Analysis Complete!")

output.register_callback('analyze_video_from_js', analyze_video_from_js)

def answer_question_from_js(question, captions, timestamps):
    try:
        answer = answer_question_realtime(question, captions, timestamps)
        send_llm_response(answer)
    except Exception as e:
        send_llm_response(f"Error generating response: {str(e)}")

output.register_callback('answer_question_from_js', answer_question_from_js)

print("\n" + "=" * 70)
print("✅ REAL AI VIDEO CHATBOT INTERFACE IS NOW WORKING!")
print("=" * 70)
print("\n🎯 To use:")
print("1. Click 'Choose Video File' in the interface")
print("2. Select any video file (MP4, AVI, MOV, MKV)")
print("3. Click 'Start REAL AI Analysis'")
print("4. Wait for analysis to complete (progress bar will show)")
print("5. Ask questions in the chat box")
print("\n⚠️ FIX APPLIED: Changed 'goggle' to 'google' in JavaScript")
```
